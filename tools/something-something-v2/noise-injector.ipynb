## Imports

```python
import json
import random
```

```python
with open('./subset_train_labels.json', 'r') as f:
    clean = json.load(f)
```

```python
clean
```

## Get the classes

```python
classes = set()

for sample in clean:
    classes.add(sample['class'])
```

```python
classes
```

## Add noise

```python
noise_level = 0.6
```

```python
noisy = clean.copy()
noise_count = 0
for sample in noisy:
    ground_truth = sample['class']
    sample['ground_truth'] = ground_truth
    
    if noise_level*len(noisy) > noise_count:
        if random.random() <= noise_level:
            options = [label for label in classes if label != ground_truth]
            sample['class'] = random.choice(options)
            noise_count += 1
```

```python
noise_count
```

## Verification

```python
mislabel_count = 0
for sample in noisy:
    if sample['class'] != sample['ground_truth']:
        mislabel_count += 1
    
mislabel_count
```

## Saving the file

```python
with open(f'subset_train_labels_{noise_level}.json', 'w') as file:
    json.dump(noisy, file)
```
