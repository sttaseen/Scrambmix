```python
import numpy as np
import matplotlib.pyplot as plt
import os
import torch
```

```python
def output_video(video_tensor, index):
    os.makedirs('output', exist_ok=True)
    for i, np_image in enumerate(video_tensor[index].squeeze().permute(1,2,3,0)):
        print(np_image.shape)
        img_array = np.array(np_image.numpy())
        plt.imshow(img_array)
        plt.axis('off')  # Turn off axis labels and ticks

        # Save the image using plt.savefig()
        output_path = f"output/{i}.jpg"
        plt.savefig(output_path, bbox_inches='tight', pad_inches=0, dpi=300)
```

## Masking

```python
A = np.random.randn(4, 1, 3, 32, 224, 224)
A.shape
```

```python
B = np.random.randn(4, 1, 3, 32, 224, 224)
B.shape
```

```python
A[..., 0, :, :, :] = 255
B[..., 1, :, :, :] = 255
```

```python
mask = np.arange(A.shape[3]) % 3 == 0
mask
```

```python
~mask
```

```python
A[:, :, :, mask, :, :] = 0
B[:, :, :, ~mask, :, :] = 0
result = A+B
```

```python
output_video(torch.tensor(A+B), 0)
```
